```python
import numpy as np
import pandas as pd
```

```python
"""
Dataset Description:
In this challenge, the dataset contains a walkatime data of past students and how they performed in
the evaluation exam. Your task is to predict the score you will get given the amount of time you
spend on coding daily.
Input : You are given one feature corresponding to time noted by walkatime.
Output : A scalar denoting the level of performance student achieved by devoting the given time.
TASK : Build a Linear Regression model on the dataset.
Upload your model, training data on your github profile.
"""
```

```python
Linear_X_Test = pd.read_csv('Linear_X_Test.csv')
Linear_X_Train = pd.read_csv('Linear_X_Train.csv')
Linear_Y_Train = pd.read_csv('Linear_Y_Train.csv')
```

```python
#our data is already splitted so we can directly apply our model
#applying Linear regression 
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression().fit(Linear_X_Train,Linear_Y_Train)
Linear_Y_Pred = regressor.predict(Linear_X_Test)
```

```python
Linear_Y_Pred
```

```python
#we can also plot
import matplotlib.pyplot as plt
plt.scatter(Linear_X_Train,Linear_Y_Train , color='red')
plt.plot(Linear_X_Train , regressor.predict(Linear_X_Train) , color = 'blue')
plt.title('Score vs amt of Time')
plt.xlabel('Amount of time')
plt.ylabel('Score')
plt.show()
```
